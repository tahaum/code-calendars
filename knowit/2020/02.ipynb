```python
import math

def is_prime(n):
    if n == 2:
        return True
    if n % 2 == 0 or n <= 1:
        return False

    sqr = int(math.sqrt(n)) + 1

    for divisor in range(3, sqr, 2):
        if n % divisor == 0:
            return False
    return True
```

```python
n_pakker_kastet = 0
pakkenr = 0
antall_pakker = 5433000

while pakkenr < antall_pakker:
    pakkenrstr = str(pakkenr)
    print(pakkenr)
    if '7' in pakkenrstr:
        for i in range(pakkenr, 0, -1):
            if is_prime(i):
                if pakkenr + i < antall_pakker:
                    n_kast = i + 1
                else:
                    n_kast = antall_pakker - i 

                n_pakker_kastet += n_kast
                pakkenr = pakkenr + n_kast - 1
                print(pakkenrstr, i, n_kast, pakkenr)
                break
                
    pakkenr += 1
    
    if pakkenr % 10**5 == 0:
        print(pakkenr)
        
        
print(antall_pakker - n_pakker_kastet)
```
